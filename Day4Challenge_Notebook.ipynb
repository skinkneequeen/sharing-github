```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
# Open/read dataset and saving it as 'rhumid'
rhumid = xr.open_dataset('rhum.mon.mean.nc')
```

```python
# Selecting the variable 'rhum' from 'rhumid' and saving it as a new dataset labelled 'rhumidity'
rhumidity = rhumid.rhum
```

```python
# First, selecting time (Fall of 1970)
# Then group by seasons
# Not sure why taking the mean
# Then selecting Fall
rhumidity_1970 = rhumidity.sel(time=slice('1970-01-01','1970-12-31')).groupby('time.season').mean(dim='time').sel(season='SON')
```

```python
# First, selecting time (Fall of 1970)
# Then group by seasons
# Not sure why taking the mean
# Then selecting Fall
rhumidity_2020 = rhumidity.sel(time=slice('2020-01-01','2020-12-31')).groupby('time.season').mean(dim='time').sel(season='SON')
```

```python
rhumidity_1970.sel(lat=0).plot(color='red',label='Fall 1970')
rhumidity_2020.sel(lat=0).plot(color='blue',label='Fall 2020')
plt.legend()
plt.title('Relative Humidity at Equator')
```

```python
rhumidity_1970.sel(lat=87.5).plot(color='black',label='Fall 1970')
rhumidity_2020.sel(lat=87.5).plot(color='darkgray',label='Fall 2020')
plt.legend(loc='lower right')
plt.title('Relative Humidity Near North Pole')
```

## Load Libraries

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import datetime
```

## Reading in surface pressure dataset

```python
# Reading in the full dataset and creating "mini datasets" based on the variables
path_name = '/Users/atoc_reu_student/Downloads/ERA5/Data/Hourly/'
file_name = 'era5_sl_hourly_Antarctic_surface_pressure_1999_nov_dec.nc'
full_dataset = xr.open_mfdataset(path_name + file_name)
e5_sp = full_dataset.sp
longitude = full_dataset.longitude
latitude = full_dataset.latitude
time = full_dataset.time
```

```python
# Setting the time variable to a simpler, more usable format

time = np.array(time) # Making the time 'dataset' into an array
time_dates = [] # Creating a blank array called 'time_dates'
for i in range(len(time)):
    string = time[i].astype(str)
    split = string.split("T", 1)
    time_dates.append(split[0])
    
```

## Setting the gridpoints of the dataset to the nearest location of Cavity and Channel camps

```python
# Locations of AWS stations at Thwaites Cavity and Channel Camps

# cavity lat = -75.033
# cavity lon = -105.617 
# channel lat = -75.050
# channel lon = -105.4334 

# Setting the gridpoints to the nearest point in reality

# Surface pressure (Pascals)
e5_sp_cavity_channel_loc_Pa = np.array(e5_sp[:,660,1018])
# Surface pressure (KiloPascals)
e5_sp_cavity_channel_loc_kPa = np.array(e5_sp[:,660,1018]) * (10 ** -3)
```

```python
##### Plotting time series of surface pressure in KiloPascals
plt.figure(figsize = (25,12))
xaxis = np.linspace(0, len(time)-1, len(time))
plt.plot(xaxis[600:1100], e5_sp_cavity_channel_loc_kPa[600:1100], '--', linewidth = 3, color = 'black', label = 'Non-AR Event')
plt.plot(xaxis[756:832], e5_sp_cavity_channel_loc_kPa[756:832], linewidth = 4, color = 'orange', label = 'AR Event')
plt.xticks(xaxis[600:1101][::24], time_dates[600:1101][::24], rotation = 40, fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Time (Days)', fontsize = 35)
plt.ylabel('Surface Pressure (kPa)', fontsize = 35)
plt.title('ERA5 Surface Pressure on Thwaites Ice Shelf\n'+'During 1999 Nov./Dec. AR Event', fontsize = 40)
plt.grid()
plt.legend(loc = 'upper right', prop = {'size':25})
plt.tight_layout()
plt.savefig("/Users/atoc_reu_student/Downloads/ERA5/Figures/Hourly/era5_surface_pressure_nov26_dec16.jpg", dpi = 100)
plt.show()
```

```python
# Plotting time series of surface pressure in Pascals
plt.figure(figsize = (25,12))
xaxis = np.linspace(0, len(time)-1, len(time))
plt.plot(xaxis[756:832], e5_sp_cavity_channel_loc_Pa[756:832], linewidth = 4, color = 'orange')
plt.xticks(xaxis[756:832][::24], time_dates[756:832][::24], rotation = 40, fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Time (Days)', fontsize = 35)
plt.ylabel('Surface Pressure (Pa)', fontsize = 35)
plt.title('ERA5 Surface Pressure on Thwaites Ice Shelf\n'+'During 1999 Nov./Dec. AR Event', fontsize = 40)
plt.grid()
plt.tight_layout()
plt.savefig("/Users/atoc_reu_student/Downloads/ERA5/Figures/Hourly/era5_surface_pressure_dec2_dec5.jpg", dpi = 100)
plt.show()
```

```python
# Plotting time series of surface pressure in kilopascals (entire time)
plt.figure(figsize = (25,12))
xaxis = np.linspace(0, len(time)-1, len(time))
plt.plot(xaxis, e5_sp_cavity_channel_loc_kPa, '--', linewidth = 3, color = 'black', label = 'Non-AR Event')
plt.plot(xaxis[756:832], e5_sp_cavity_channel_loc_kPa[756:832], linewidth = 4, color = 'orange', label = 'AR Event')
plt.xticks(xaxis[::100], time_dates[::100], rotation = 40, fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Time (Days)', fontsize = 35)
plt.ylabel('Surface Pressure (kPa)', fontsize = 35)
plt.title('ERA5 Surface Pressure on Thwaites Ice Shelf\n'+'During 1999 Nov./Dec.', fontsize = 40)
plt.grid()
plt.legend(loc = 'upper right', prop = {'size': 25})
plt.tight_layout()
plt.savefig("/Users/atoc_reu_student/Downloads/ERA5/Figures/Hourly/era5_surface_pressure_nov1_dec29.jpg", dpi = 100)
plt.show()
```
